```python
import leafmap.foliumap as leafmap
```

```python
m = leafmap.Map(center=[36.221834032579274, 136.92268626950698], zoom=16)  # center=[lat, lon]
m.add_basemap("HYBRID")
gsi_std_url = 'https://cyberjapandata.gsi.go.jp/xyz/std/{z}/{x}/{y}.png' 
m.add_wms_layer(
    url=gsi_std_url, layers='0', name='GSI Standard', format='image/png', shown=True
)
gifu_cs_url = 'https://kenzkenz.xsrv.jp/tile/gihu/cs/{z}/{x}/{-y}.png'
m.add_wms_layer(
    url=gifu_cs_url, layers='0', name='Gifu CS Map by Hinata GIS', format='image/png', shown=True
)

in_geojson = 'https://storage.googleapis.com/prism-drr/shared/leaflet_inoue_maps/geojson/kiunLS.geojson'
m.add_geojson(
    in_geojson, layer_name="1586 Landlides",fill_colors=['red', 'yellow', 'green', 'orange']
)
m
```

```python
m.to_html("kiunjo.html")
```
